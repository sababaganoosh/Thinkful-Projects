```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
#TODO: import file
df = pd.read_csv(r'C:\Users\Avraham\Desktop\Python\economicdata2016-2016.csv')
plt.figure(figsize=(10, 5))
#TODO: Create plot settings

plt.subplot(1, 2, 1)
plt.scatter(x = df['Integrity of the legal system'], y = df['Reliability of police'], color='green')
plt.ylabel('Police Reliability')
plt.title('Integrity of Legal System vs Police Reliability')

plt.subplot(1, 2, 2)
plt.scatter(x = df['Freedom to trade internationally'], y = df['Freedom of foreigners to visit'], color='blue')
plt.ylabel('Freedom of foreigners to visit')
plt.title('Freedom of international trade vs Freedom of foreigners to visit')


plt.show()
# There is a positive linear coorelation between Police Reliability and Integrity of the legal system.
# There is a positive exponential coorelation between Integrity of the legal system and Freedom of foreigners to visit
```
