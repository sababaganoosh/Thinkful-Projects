```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#Load Data set
df = pd.read_csv(r'C:\Users\bonda\OneDrive\Desktop\StudentsPerformance.csv')

# Creating variables for each of the four conditions to graph using the basic
# matplotlib functionality.
Male_study_math_score = df.loc[(df['gender']=='male')&(df['test preparation course']=='completed'),'math score']
Female_study_math_score = df.loc[(df['gender']=='female')&(df['test preparation course']=='completed'),'math score']
Male_no_study_math_score = df.loc[(df['gender']=='male')&(df['test preparation course']=='none'),'math score']
Female_study_math_score = df.loc[(df['gender']=='female')&(df['test preparation course']=='none'),'math score']

# Making overlapping histograms.
plt.hist(Male_study_math_score, color='red',  alpha=.5,label='Boys that study')  # alpha just controls the opacity
plt.hist(Female_study_math_score, color='blue', alpha=.5,label='Girls that study')
plt.hist(Male_no_study_math_score, color='green', alpha=.5,label="Boys that don't study")  
plt.hist(Female_study_math_score, color='yellow', alpha=.5,label="Girls that don't study")
plt.xlabel('Math Score')
plt.legend(loc='upper right')
plt.title('Plot 1: Math scores by gender and study practices')
plt.show()

# Showing the  information in a seaborn facet grid.
sns.set(style="ticks")  #Setting the overall aesthetic

# Tell seaborn about the structure of our data.
g = sns.FacetGrid(df, row="gender", col="test preparation course")
# Name the plot type and the variable to be plotted using the structure.
g.map(plt.hist, "math score", color="steelblue",  lw=0)
# Moving the plots apart to make room for our titles.
plt.subplots_adjust(top=0.9)
# Making a more informative axis name.
g.set_axis_labels('Math Score')
plt.suptitle('Plot 2: Math Score by gender and study practices')
# Removing excess lines around the plot.
sns.despine(trim=True)
plt.show()

# Comparing groups using boxplots.
ax = sns.boxplot(x='test preparation course',y='math score',hue='gender',data=df)  
plt.title('Plot 3: Math Score by gender and study practices')
sns.despine(offset=10, trim=True)
ax.set(xlabel='', ylabel='Math Scores')
plt.show()

# Setting the overall aesthetic.
sns.set(style="whitegrid")

g = sns.factorplot(x="gender", y="math score", hue="test preparation course", data=df,
                   size=6, kind="point", palette="pastel",ci=95,dodge=True,join=False)
g.despine(left=True)
g.set_ylabels("Math Score")
g.set_xlabels("")
plt.title('Pointplot: Math Score by gender and study practices')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#Load Data set
df = pd.read_csv(r'C:\Users\bonda\OneDrive\Desktop\StudentsPerformance.csv')

g = sns.lmplot(y='math score', # Variable 1.
               x='writing score', # Variable 2.
               data=df, # Data
               fit_reg=False, # If set to true, plots a regression line.
               scatter_kws={'alpha':0.4}) # Set points to semi-transparent to see overlaping points.
g.set_ylabels("Math Score")
g.set_xlabels("Writing Score")
plt.title('Scatterplot: Math Score by Writing Score')
plt.show()

g = sns.lmplot(y='math score', 
               x='writing score',
               data=df, # Data.
               fit_reg=True, # The regression line also includes a 95% confidence envelope.
               scatter_kws={'alpha':0.4})
g.set_ylabels("Math Score")
g.set_xlabels("Writing Score")
plt.title('Scatterplot with regression line: Math Score by Writing Score')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#Load Data set
df = pd.read_csv(r'C:\Users\bonda\OneDrive\Desktop\StudentsPerformance.csv')

# Setting the overall aesthetic.
sns.set(style="white")

g = sns.lmplot(y='math score', 
               x='writing score',
               hue='gender',
               data=df, # Data.
               col='test preparation course',
               fit_reg=True,
               ci=False,
               scatter_kws={'alpha':0.4})
g.set_ylabels("Math Score")
g.set_xlabels("Writing Score")
plt.suptitle('Scatterplot Regression: Math Score by Writing Score, split by gender and study habits')
plt.subplots_adjust(top=0.9)
plt.show()

g = sns.lmplot(y='math score', 
               x='writing score',
               hue='test preparation course',
               data=df, # Data.
               col='gender',
               fit_reg=True, 
               ci=False,
               scatter_kws={'alpha':0.4}, 
               palette='pastel',
               lowess=True) # Using local weighing to fit a line.
g.set_ylabels("Math Score")
g.set_xlabels("Writing Score")
plt.suptitle('Scatterplot Lowess: Math Score by Writing Score, split by study habits and gender')
plt.subplots_adjust(top=0.9)
plt.show()
```
